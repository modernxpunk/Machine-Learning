```python
import pandas as pd

from sklearn.metrics import accuracy_score, confusion_matrix, plot_confusion_matrix, classification_report
from sklearn.model_selection import train_test_split
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis  
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier  
from sklearn.naive_bayes import GaussianNB  
from sklearn.tree import DecisionTreeClassifier  
from sklearn.svm import SVC

import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('practice1_data.csv')
```

```python
data
```

```python
features = ['x1', 'x3', 'x5', 'x7', 'x10']
```

```python
X = data[features]
y = data['classes']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.8, random_state=10)
```

```python
model_LinearDiscriminantAnalysis = LinearDiscriminantAnalysis()
model_KNeighborsClassifier = KNeighborsClassifier(n_neighbors=10)
model_GaussianNB = GaussianNB()
model_DecisionTreeClassifier = DecisionTreeClassifier()
model_SVC = SVC()
model_LogisticRegression = LogisticRegression()
```

```python
model_LinearDiscriminantAnalysis.fit(X_train, y_train)
model_KNeighborsClassifier.fit(X_train, y_train)
model_GaussianNB.fit(X_train, y_train)
model_DecisionTreeClassifier.fit(X_train, y_train)
model_SVC.fit(X_train, y_train)
model_LogisticRegression.fit(X_train, y_train)
```

```python
ldAnalysis = tuple(np.array(model_LinearDiscriminantAnalysis.predict(X_test)))
KNN = tuple(np.array(model_KNeighborsClassifier.predict(X_test)))
Gaussian = tuple(np.array(model_GaussianNB.predict(X_test)))
dTree = tuple(np.array(model_DecisionTreeClassifier.predict(X_test)))
SVC = tuple(np.array(model_SVC.predict(X_test)))
lRegression = tuple(np.array(model_LogisticRegression.predict(X_test)))
```

```python
percent_of_correct_answer = {ldAnalysis: 0, KNN: 0, Gaussian: 0, dTree: 0, SVC: 0, lRegression: 0}

for model in percent_of_correct_answer:
    count_corrent_classes = np.size(np.where(model == y_test))
    percent_of_correct_answer[model] = count_corrent_classes / len(y_test)

print(percent_of_correct_answer.values())
```

```python
fig, ax = plt.subplots(dpi = 120)
plot_confusion_matrix(model_LinearDiscriminantAnalysis, X_test, y_test, ax = ax)
```
