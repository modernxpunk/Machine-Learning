```python
import pandas as pd

from sklearn.discriminant_analysis import LinearDiscriminantAnalysis  
from sklearn.neighbors import KNeighborsClassifier  
from sklearn.naive_bayes import GaussianNB  
from sklearn.tree import DecisionTreeClassifier  
from sklearn.svm import SVC

import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('practice1_data.csv')
```

```python
data
```

```python
features = ['x1', 'x2', 'x3', 'x4', 'x5', 'x6', 'x7', 'x8', 'x9', 'x10']
```

```python
X = data[features]
y = data['classes']
```

```python
model_LinearDiscriminantAnalysis = LinearDiscriminantAnalysis()
model_KNeighborsClassifier = KNeighborsClassifier()
model_GaussianNB = GaussianNB()
model_DecisionTreeClassifier = DecisionTreeClassifier()
model_SVC = SVC()
```

```python
model_LinearDiscriminantAnalysis.fit(X, y)
model_KNeighborsClassifier.fit(X, y)
model_GaussianNB.fit(X, y)
model_DecisionTreeClassifier.fit(X, y)
model_SVC.fit(X, y)
```

```python
original = y
sample = len(original)
```

```python
ldAnalysis = tuple(np.array(model_LinearDiscriminantAnalysis.predict(X)))
KNN = tuple(np.array(model_KNeighborsClassifier.predict(X)))
Gaussian = tuple(np.array(model_GaussianNB.predict(X)))
dTree = tuple(np.array(model_DecisionTreeClassifier.predict(X)))
SVC = tuple(np.array(model_SVC.predict(X)))
```

```python
percent_of_correct_answer = {ldAnalysis: 0, KNN: 0, Gaussian: 0, dTree: 0, SVC: 0}

for model in percent_of_corrent_answer:
    count_corrent_classes = np.size(np.where(model == original))
    percent_of_correct_answer[model] = count_corrent_classes / sample

print(percent_of_correct_answer.values())
```
