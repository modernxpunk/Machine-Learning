```python
from sklearn.datasets import make_regression
from sklearn.tree import DecisionTreeRegressor
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
X_test, y_test = make_regression(n_features=1, noise=9)
plt.scatter(X_test,y_test)
plt.show()
```

```python
# Selecting Data for Modeling
boston_houses_data = pd.read_csv('boston.csv')
boston_houses_data

# Input features in order:
# 1) CRIM: per capita crime rate by town
# 2) ZN: proportion of residential land zoned for lots over 25,000 sq.ft.
# 3) INDUS: proportion of non-retail business acres per town
# 4) CHAS: Charles River dummy variable (1 if tract bounds river; 0 otherwise)
# 5) NOX: nitric oxides concentration (parts per 10 million) [parts/10M]
# 6) RM: average number of rooms per dwelling
# 7) AGE: proportion of owner-occupied units built prior to 1940
# 8) DIS: weighted distances to five Boston employment centres
# 9) RAD: index of accessibility to radial highways
# 10) TAX: full-value property-tax rate per $10,000 [$/10k]
# 11) PTRATIO: pupil-teacher ratio by town
# 12) B: The result of the equation B=1000(Bk - 0.63)^2 where Bk is the proportion of blacks by town
# 13) LSTAT: % lower status of the population

# Output variable:
# 1) MEDV: Median value of owner-occupied homes in $1000's [k$]
```

```python
# Selecting The Prediction Target
y = boston_houses_data['MEDV']
```

```python
# Choosing features
boston_features = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX',
       'PTRATIO', 'B', 'LSTAT']
```

```python
X = boston_houses_data[boston_features]
```

```python
X.head()
```

```python
# Building the Model
boston_model = DecisionTreeRegressor(random_state=1)
boston_model.fit(X, y)
```

```python
y.head()
```

```python
boston_model.predict(X.head())

# Prediction has been successes
```

```python
# Draw graph

origin_medv = y
predicted_medv = boston_model.predict(X)

plt.scatter(origin_medv, predicted_medv)
```
